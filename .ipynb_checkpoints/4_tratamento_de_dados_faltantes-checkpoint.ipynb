# Tratando os dados: removendo dados null, None, NaN.

```python
import pandas as pd
```

### Identificando os dados e suas quantidades

```python
dados = pd.read_csv('arquivos/dados_residenciais.csv', sep =';')
dados.info()
```

## Identificando e removendo

```python
dados[dados['Valor'].isnull()]
```

```python
#Removendo os NaN usando dropna()
dados.dropna(subset = ['Valor'], inplace = True)
```

## Tratando IPTU e condomínio em apartamentos

### Isolando os apartamentos com condomínio NaN

```python
#selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
#dados = dados[~selecao]
dados = dados[~selecao]
```

```python
dados
```

```python
dados.fillna({'Condominio': 0, 'IPTU': 0}, inplace = True)
```
