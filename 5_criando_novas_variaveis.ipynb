## Criar variáveis de valor bruto, m². Agrupar 'casas' apenas

```python
import pandas as pd
```

```python
dados_cru = pd.read_csv('arquivos/dados_residenciais_limpos.csv', sep=';')
dados_cru.head()
```

```python
dados_cru['Valor bruto'] = dados_cru['Valor'] + dados_cru['IPTU'] + dados_cru['Condominio']
```

```python
dados_cru['Valor m2'] = (dados_cru['Valor'] / dados_cru['Area']).round(2)
```

```python
dados_cru.head()
```

### Agora que foram criadas as novas variáveis, vamos agrupar as 'casas' em 'Casa'

```python
dados_cru['Tipo'].unique()
```

```python
casa = []
for item in dados_cru['Tipo'].unique():
    if(item == 'Casa' or item == 'Casa de Vila' or item == 'Casa de Condomínio'):
        casa.append(item)
```

```python
# dados_raw['Tipo agregado'] = dados_raw['Tipo'].apply(lambda x: 'casa' if x in casa else 'Apartamento')
dados_cru['Tipo agregado'] = dados_cru['Tipo'].apply(lambda x: 'casa' if x in casa else 'Apartamento')
```

```python
dados_cru.head()
```

```python
#Agora vou atualizar o arquivo original com os dados novos. 
dados_cru.to_csv('arquivos/dados_residenciais_limpos.csv', sep=';', index = False)
```
